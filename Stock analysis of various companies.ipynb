```python
7• Add a new column ‘Trend’ whose values are based on the 'Daily Return'.
```

```python
def trend(x):
  if x > -0.015 and x <= 0.015:
    return 'Slight or No change'
  elif x > 0.015 and x <= 0.04:
    return 'Slight Positive'
  elif x < -0.015 and x >= -0.4:
    return 'Slight Negative'
  elif x > 0.04 and x <= 0.06:
    return 'Positive'
  elif x < -0.04 and x >= -0.06:
    return 'Negative'
  elif x > 0.06 and x <= 0.07:
    return 'Among top gainers'
  elif x < -0.06 and x >= -0.07:
    return 'Among top losers'
  elif x > 0.07:
    return 'Bull run'
  elif x <= -0.07:
    return 'Bear drop'

amazon['Trend']= pd.Series(np.zeros(amazon['Daily Return'].count()))
amazon['Trend']= amazon['Daily Return'].apply(lambda x:trend(x))
amazon.head()
```

```python
m_dict_1=dict()
plt.figure(figsize=(8,8))

for Trend in set(amazon.Trend):
    m_dict_1[Trend]=amazon[amazon.Trend==Trend].Trend.count()
plt.pie(m_dict_1.values(),labels=m_dict_1.keys(), autopct='%1.1f%%')
plt.show()
```

```python
9• What was the correlation between the daily returns of different stocks?
```

```python
import pandas_datareader.data as web
end = datetime.now()
start_date = datetime(end.year-1, end.month, end.day)
combined_df = web.DataReader(['AMZN', 'GOOG', 'MSFT', 'AAPL'],
                            'yahoo', start = start_date, end=end)['Adj Close']
```

```python
combined_df.head()
```

```python
combined_df.dropna(inplace = True, axis = 0)
combined_df.head()
```

```python
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = 'any', axis = 0)
pct_chg_df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(10,5))
sns.heatmap(pct_chg_df.corr(), annot=True)
```
